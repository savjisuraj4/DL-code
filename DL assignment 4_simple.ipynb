```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input,Dense
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
```

```python
dataset=pd.read_csv("creditcard.csv")
```

```python
dataset
```

```python
x=dataset.drop("Class",axis=1)
y=dataset["Class"]
```

```python
x_scaled=StandardScaler().fit_transform(x.values)
print(x.values)
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x_scaled,y,test_size=0.2,random_state=2021)  
#train_data,test_data,train_labels,test_labels means Xtrain,Xtest,Ytrain,Ytest
```

```python
input_dim=Xtrain.shape[1]
```

```python
input_layer=Input(shape=input_dim)
```

```python
encode_dim=14
encoded=Dense(encode_dim,activation="relu")(input_layer)
```

```python
decoded=Dense(input_dim,activation="relu")(encoded)
```

```python
autoencoder=tf.keras.models.Model(input_layer,decoded)
```

```python
autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.2),
    loss='mean_squared_error',
    metrics=['accuracy']
)
```

```python
autoencoder.fit(Xtrain,Ytrain,epochs=10,batch_size=256,shuffle=True,validation_split=0.1)
```

```python
reconstructed=autoencoder.predict(Xtest)
```

```python
mse=np.mean(np.power(Xtest-reconstructed,2),axis=1)
```

```python
mse_threshold=np.percentile(mse,99) #Calculate the 99th percentile
print(mse_threshold)
```

```python
anomalies=mse>mse_threshold
print(f"Number of anomalies samples: {np.sum(anomalies)}")
print(f"Indices of anomalies samples: {np.where(anomalies)[0]} ")
```
